# 10 .groupby():

Allows you to group data based on specific columns and perform aggregate operations on the grouped data.

Let's see an example:

```python
# Sample sales data
data = {
    'Product': ['Product A', 'Product B', 'Product A', 'Product B', 'Product A'],
    'Category': ['Electronics', 'Clothing', 'Electronics', 'Clothing', 'Electronics'],
    'Price': [500, 50, 600, 40, 550],
    'Quantity': [2, 3, 1, 4, 2]
}

# Create a DataFrame
df = pd.DataFrame(data)

df.head()
```

```python
# Group by "Category" and calculate aggregate values
grouped = df.groupby('Category').agg(
    total_sales=pd.NamedAgg(column='Price', aggfunc='sum'),
    average_price=pd.NamedAgg(column='Price', aggfunc='mean')
).reset_index()

# We use .reset_index() to reset the index and make the "Category" column a regular column.

grouped.head()
```

The resulting grouped DataFrame shows the total sales and average price for each category. This is just a simple example; .groupby() can be used with more complex data and multiple columns to perform various aggregations on grouped data.

# 13 .merge() and .join():

These methods are used to combine multiple DataFrames based on common columns or indices.

```python
# Using .merge()

# Sample student data
student_data = {
    'StudentID': [1, 2, 3, 4, 5],
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Emily'],
    'Age': [20, 22, 21, 19, 20]
}

# Sample test scores data
scores_data = {
    'StudentID': [1, 2, 3, 4, 5],
    'TestScore': [85, 78, 92, 70, 88]
}

# Create DataFrames
students_df = pd.DataFrame(student_data)
scores_df = pd.DataFrame(scores_data)

# Merge DataFrames based on "StudentID"
merged_df = pd.merge(students_df, scores_df, on='StudentID')

merged_df.head()
```

```python
# Using .join()


# Sample student data
student_data = {
    'StudentID': [1, 2, 3, 4, 5],
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Emily'],
    'Age': [20, 22, 21, 19, 20]
}

# Sample enrollment data
enrollment_data = {
    'StudentID': [1, 2, 3, 4, 5],
    'Course': ['Math', 'History', 'Science', 'Math', 'English']
}

# Create DataFrames
students_df = pd.DataFrame(student_data)
enrollment_df = pd.DataFrame(enrollment_data)

# Join DataFrames based on "StudentID"
joined_df = students_df.join(enrollment_df.set_index('StudentID'), on='StudentID')

joined_df.head()
```

In both examples, the DataFrames are combined based on a common column, such as "StudentID". The choice between .merge() and .join() depends on your specific use case and requirements. .merge() offers more flexibility for different types of joins and customization, while .join() is a more simplified method for combining DataFrames based on indices.

# 14 .pivot_table():

Creates a pivot table from a DataFrame, allowing you to summarize and aggregate data based on a specific column.

```python
# Sample sales data
data = {
    'Product': ['Product A', 'Product B', 'Product A', 'Product B', 'Product A'],
    'Category': ['Electronics', 'Clothing', 'Electronics', 'Clothing', 'Electronics'],
    'Quarter': ['Q1', 'Q1', 'Q2', 'Q2', 'Q3'],
    'Sales': [1000, 500, 800, 600, 1200]
}

# Create a DataFrame
df_sales = pd.DataFrame(data)

df_sales.head()
```

```python
# Create a pivot table to summarize sales by category and quarter
pivot_table = df_sales.pivot_table(index='Category', columns='Quarter', values='Sales', aggfunc='sum', fill_value=0)

pivot_table.head()
```

# 15 .fillna():

Fills missing values in the DataFrame with specified values or methods.

```python
# Sample student data with missing age values
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Emily'],
    'Age': [20, np.nan, 21, np.nan, 22]
}

# Create a DataFrame
df_nan = pd.DataFrame(data)

# Calculate the average age
average_age = df_nan['Age'].mean()

# Fill missing age values with the average age
df_nan['Age'] = df_nan['Age'].fillna(average_age)

df_nan.head()
```

![green-divider](https://user-images.githubusercontent.com/7065401/52071924-c003ad80-2562-11e9-8297-1c6595f8a7ff.png)

# 16 .drop_duplicates():

Removes duplicate rows based on specified columns.

```python
# Sample sales data with duplicate entries
data = {
    'Product': ['Product A', 'Product B', 'Product A', 'Product B', 'Product A'],
    'Category': ['Electronics', 'Clothing', 'Electronics', 'Clothing', 'Electronics'],
    'Price': [500, 50, 600, 40, 550]
}

# Create a DataFrame
df_duplicates = pd.DataFrame(data)

df_duplicates.head()
```

```python
# Remove duplicate rows based on "Product" column
df_no_duplicates = df_duplicates.drop_duplicates(subset='Product')

df_no_duplicates.head()
```

These are just a few examples of the many methods available in Pandas for data analysis and manipulation. Each method serves a specific purpose and allows you to efficiently work with tabular data.
